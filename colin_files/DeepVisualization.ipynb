### Loading Function

```python
def consolidate(fileformat):
    files = np.sort(np.array(glob.glob(fileformat)))
    tensors=[]
    for fname in files:
        with open(fname,'rb') as f:
            tensors.append(np.load(f))
    return np.vstack(tensors)


def loaddata(loaddir):
    fmt=loaddir+"group_*/"
    msk=consolidate(fmt+'attn/*.npy')
    etens_t=consolidate(fmt+'entropy_t/*.npy')
    etens_c=consolidate(fmt+'entropy_c/*.npy')
    ctens_c=consolidate(fmt+'codelength_c/*.npy')
    ctens_t=consolidate(fmt+'codelength_t/*.npy') 
    ptens=consolidate(fmt+'pleak/*.npy')
    ltens=consolidate(fmt+'tlens/*.npy')
    ttens=consolidate(fmt+'tok/*.npy')

    return {'ec': etens_c,
            'et': etens_t,
            'cc': ctens_c,
            'ct': ctens_t,
            'p': ptens,
            'm': msk,
            't': ttens,
            'l': ltens,
            'dir': loaddir}

def takeaverage(data):

    nos=data['m'].sum(axis=0)
    weights=data['m']/data['m'].sum(axis=0,keepdims=True)
    et=(data['et']*weights).sum(axis=0)
    ct=(data['ct']*weights[:,:-1]).sum(axis=0)

    positions=data['l'].cumsum(axis=-1)
    pflat = positions.flatten()
    msk = data['m'].flatten()
    gmask=np.ones_like(data['m'])
    gmask[:,-1]=0
    gmask=gmask.astype('bool').flatten()
    validpos, inv = np.unique(pflat, return_inverse=True)
    nsamples = np.bincount(inv, weights=msk)
    #print(nsamples)
    e_sum = np.bincount(inv, weights=data['ec'].flatten() * msk)
    ec = np.divide(e_sum, nsamples, out=np.zeros_like(e_sum), where=nsamples>0)
    c_sum = np.bincount(inv[gmask], weights=data['cc'].flatten() * msk[gmask],minlength=nsamples.shape[0])
    cc = np.divide(c_sum, nsamples, out=np.zeros_like(c_sum), where=nsamples>0)
    
    averages={"nt": nos,
              'ct': ct,
              'et': et,
              "nc": nsamples,
              'vp': validpos,
              'ec': ec,
              'cc': cc,
              }

    return averages
```
